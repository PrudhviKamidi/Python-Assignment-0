```python
import re

# Registartion function
    
def registration():
    temp=0
    
    pattern = '^[a-z]+[\._]?[a-z 0-9]+[@]\w+[.]\w{2,3}$'
    

    

    # validating email


    username=input('Enter valid Email (eg:asdef@guvi.com) :')

    with open('data.txt','r') as d2:

        if re.search(pattern, username):

            for line in d2:

                if username in line:

                    print("Username with this email already Registered")
                    temp=1
                    d2.close()
                    break
        else:
            print('Enter valid email address')
            temp=1
        

    # validating password
    
    specials = ['@','!','#','^','%','$','&','*','(',')']

    if temp != 1:
        print("\n Password length between 5-16 ,Must have minimum \n one special character,\n one digit,\n one uppercase,\n one lowercase characters")
        password=input('Enter password :')
        if len(password) <16 and len(password) >5:

            if re.search('[0-9]',password):

                if re.search('[A-Z]',password):

                    if re.search('[a-z]',password):

                        if any(char in specials for char in password):

                            with open('data.txt', 'a') as f:

                                f.write(f"\n{username},{password}")
 
                                print("-- Registration is Successful --")
                                
                        else:
                            print("-- Password should contain atleast one special character --")
                    else:
                        print("-- Password should contain atleast one lowercase letter --")

                else:
                    print("-- Password should contain atleast one uppercase letter --")    
            else:
                print("-- Password should contain atleast one Digit --")

        else:
            print('-- password length must be between 5 to 16 --')

        
# forgot password definition

def forgot_password():
    
    forgot_mail=input('Enter the registered email: \n')

    with open('data.txt','r') as d3:

        for line in d3:

            if forgot_mail in line:
                pas=line.split(',')

                print("The Registered password is",pas[-1])
                d3.close()
                break
        else:
            print("Email is not valid")


# login function definition

def login():
    login_mail=input('Login E-mail:')
    login_pass=input('Login Password:')

    with open('data.txt','r') as d1:
        
        for line in d1:
            
            if login_mail and login_pass in line:
                
                # removing escape sequence from line
                escape = re.sub(r"(\\n|\\r|\\t|\\)", "", line).strip()             
                check=escape.split(',')
                
                if login_mail ==check[0] and login_pass == check[-1]:
                    print("Login Sucessful")
                    break

        else:
                new_input=int(input("\n Incorrect Details ,select: \n 1-For Registration \n 2-Forgot Password \n 3-Retry Login \n TO EXIT  (enter any number, other than 1,2,3.) \n"))

                if new_input == 1:
                    registration()
                elif new_input == 2:
                    forgot_password()
                elif new_input == 3:
                     login()

                else:
                    print("-- Register to continue --")
                    
# Program input

user_input=int(input("-- Enter the option-- \n 1-Registration \n 2-Login \n 3-Forget Password \n TO EXIT  (enter any number, other than 1,2,3.) \n"))
new_input =0
if user_input == 1:
    registration()
elif user_input == 2:
    login()
elif user_input == 3:
    forgot_password()
else:
    print("-- THANK YOU FOR VISITING --")
```
